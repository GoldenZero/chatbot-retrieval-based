```python
import jieba.posseg as pseg
import codecs
from gensim import corpora, models, similarities
```

```python
result = []
filename = 're_build_test_question_cut_9keys.csv'
with open(filename, 'r') as f:
    text = f.readlines()

        
```

```python
import time
len_text  = len(text)
resutl = []
count =0
for line in text:
    result.append(line.split(" "))
corpus = result
```

```python
dictionary = corpora.Dictionary(corpus)
```

```python
import pandas as pd
df = pd.read_csv('re_build_test_question_cut_9keys.csv')
textlines = []
for j in range(len(df)):
    text = df['source_question'].values[j].split(" ")
    textlines.append(dictionary.doc2bow(text))
# print(pseg.cut(df['Ground_Truth'][0]))
```

```python
doc_vectors = textlines
print(len(doc_vectors))
# print(doc_vectors)
tfidf = models.TfidfModel(doc_vectors)
tfidf_vectors = tfidf[doc_vectors]
print(len(tfidf_vectors))
```

```python
import numpy as np
key_list = ['Ground_Truth','e0','e1','e2','e3','e4','e5','e6','e7','e8']
index = similarities.MatrixSimilarity(tfidf_vectors)
count = 0
time_count = 0
start = time.time()
count = {}
for i in range(10):
    count[i] = 0
for i in range(len(df[0:10000])):
    result_list = []
    for testkey in key_list:
        query = df[testkey].values[i].split(" ")
        query_bow = dictionary.doc2bow(query)
        sims = index[query_bow]
        result_list.append(sims[i])
    for i in range(10):
        if 0 in np.array(result_list).argsort()[-(i+1):]:
            count[i]+=1
    time_count+=1
    if time_count%10 == 0:
        print(len(df[0:10000])/time_count*(time.time()-start)-(time.time()-start))
        
```

```python
for i in count:
    print(i+1,count[i]/len(df[0:10000]))
```
